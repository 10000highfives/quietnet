```python
import math
import struct
import pyaudio

CHUNK = 1024
CHANNELS = 1
RATE = 44100
FORMAT = pyaudio.paInt16
WIDTH = 2

SAMPLES = 5

def capture_frames(samples):
    p = pyaudio.PyAudio()

    frames = []
    
    stream = p.open(format=FORMAT,
                    channels=CHANNELS,
                    rate=RATE,
                    input=True,
                    frames_per_buffer=CHUNK)
    
    for i in range(0, samples):
        data = stream.read(CHUNK)
        frames.append(data)
    
    stream.stop_stream()
    stream.close()
    p.terminate()
    
    return frames

def unpack_frame(frame):
    return [struct.unpack('h',x)[0] for x in frame]

def chunks(l, n):
    for i in xrange(0, len(l), n):
        yield l[i:i+n]
```

```python
frame = b''.join(capture_frames(4))
frame = unpack_frame(list(chunks(frame, 2)))
```

```python
import pyaudio
import quietnet

CHUNK = 1024
CHANNELS = 1
RATE = 44100
FORMAT = pyaudio.paInt16
WIDTH = 2

SAMPLES = 5

def capture_buffers(samples, skip=2):
    p = pyaudio.PyAudio()

    buffers = []
    
    stream = p.open(format=FORMAT,
                    channels=CHANNELS,
                    rate=RATE,
                    input=True,
                    frames_per_buffer=CHUNK)
    
    # ignore the first chunk as it is usually weird
    for i in range(0, skip):
        data = stream.read(CHUNK)
            
    for i in range(0, samples):
        data = stream.read(CHUNK)
        buffers.append(data)
    
    stream.stop_stream()
    stream.close()
    p.terminate()
    
    return buffers
```

```python
# capture 300 buffers of 1024 bits at 44100 sample rate so almost 7 seconds.
buffers = capture_buffers(300)
buffers = [quietnet.unpack(b) for b in buffers]
```

```python
for sample in buffers:
    plot(sample)
```

```python
# plot the fft of every buffer
ffts = [quietnet.fft(b) for b in buffers]

for fft in ffts:
    plot(fft)
```

```python
#Focus in on the frequency of interest
search_freq = 18000
peak = quietnet.get_peak(search_freq, RATE, CHUNK)

for fft in ffts:
    plot(fft[peak-3:peak+4])
```

```python
# figure out which ffts have the frequency

for fft in ffts:
    has_freq = quietnet.has_freq(fft, search_freq, RATE, CHUNK)
    if has_freq:
        plot(fft[peak-3:peak+4], 'green')
    else:
        plot(fft[peak-3:peak+4], 'red')
```

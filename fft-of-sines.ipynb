```python
import math
import numpy as np

CHUNK = 1024
CHANNELS = 1
RATE = 44100
WIDTH = 2

def synth_complex(freq=[440], coef=[1], datasize=10000):
    amp=8000.0 
    sine_list=[]
    for x in range(datasize):
        samp = 0
        for k in range(len(freq)):
            samp = samp + coef[k] * math.sin(2*math.pi*freq[k]*(x/float(RATE)))
        sine_list.append(int(samp*amp/2))
    return sine_list
```
